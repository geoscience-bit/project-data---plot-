E1 solution

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
log_data = pd.read_csv('/content/drive/My Drive/project/log_data_csv.csv')
log_data.head()
```

```
#Plotting Neutron Porosity vs Depth based on Gamma Ray range
plt.figure(figsize=(10,5))
plt.scatter(x=log_data['NPHI'],y=log_data['DEPTH'], s=20, c=log_data['GR'], cmap='inferno_r')
plt.xlabel('Neutron Porosity', fontsize=15)
plt.ylabel('Depth', fontsize=15)
plt.title("Neutron Porosity vs Depth", fontsize=15)
plt.gca().invert_yaxis()
c=plt.colorbar()
c.ax.set_title('Gamma Ray Scale')
```
